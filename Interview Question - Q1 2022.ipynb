```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_color(rgb):
    return '{' + f'{rgb[0]}, {rgb[1]}, {rgb[2]}' + '}'

def get_row(row):
    return ('{{ ' + ", ".join([get_color(c) for c in row]) + ' }}')
    
def get_img(img):
    return ('{{\n ' + ",\n ".join([get_row(r) for r in img]) + ' \n}}')
    
def get_prefix(img):
    return f'std::array<std::array<color, {img.shape[1]}>, {img.shape[0]}> input ='

def get_1d_prefix(img):
    return f'std::array<color, {img.shape[0]}> input ='
```

```python
def get_1d_colors(base_color, covarince, height):
    cov = np.eye(3,3) * np.array(covarince)
    return np.random.multivariate_normal(np.array(base_color), cov, height)
```

```python
def get_colors(base_color, covarince, height, width):
    cov = np.eye(3,3) * np.array(covarince)
    return np.random.multivariate_normal(np.array(base_color), cov, (height, width))
```

# 1D array

```python
patch_height = 10
stick_var = [4,5,10]
patch_var = [4,4,4]
white_base = [229, 237, 249]
patch_base = [124, 170, 70]
before_patch = get_1d_colors(white_base, stick_var, 30)
patch = get_1d_colors(patch_base, patch_var, patch_height)
after_patch = get_1d_colors(white_base, stick_var, 10)
img = np.vstack([before_patch, patch, after_patch])
display_img = np.tile(img, 5).reshape((-1, 5, 3))
plt.imshow(display_img / 255)
```

```python
print(get_1d_prefix(img), get_row(img) + ";")
```
